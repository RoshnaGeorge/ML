A6: Cosine Similarity Measure

```python
import pandas as pd
import numpy as np

# Load -> thyroid dataset
thyroid_data = pd.read_excel(
    "Lab Session Data.xlsx",
    sheet_name="thyroid0387_UCI"
)

print(thyroid_data.head(2))
```

complete vectors

```python
first_2_ob = thyroid_data.iloc[:2].copy()
first_2_ob = first_2_ob.replace('?', np.nan)
en_data = pd.get_dummies(first_2_ob, dummy_na=True)
en_data = en_data.astype(float)
en_data = en_data.fillna(0)
vector_A = en_data.iloc[0].to_numpy()
vector_B = en_data.iloc[1].to_numpy()
print("vector A:")
print(vector_A)
print("\nvector B:")
print(vector_B)
print("\nlen of vector A:", len(vector_A))
print("len of vector B:", len(vector_B))
```

cosine similarity

```python
dot_prod = np.dot(vector_A, vector_B)
mag_A = np.linalg.norm(vector_A)
mag_B = np.linalg.norm(vector_B)
cos_sim = dot_prod / (mag_A * mag_B)
print("dot prod of A and B:", dot_prod)
print("magnitude of vector A:", mag_A)
print("magnitude of vector B:", mag_B)
print("cosine similarity:", cos_sim)
```
