A4: Data Exploration of Thyroid Dataset


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

loading the data

```python
# Load the thyroid0387_UCI worksheet
thyroid_data = pd.read_excel(
    "Lab Session Data.xlsx",
    sheet_name="thyroid0387_UCI"
)

# Display the first five rows
print(thyroid_data.head())
```

study each attribute and values


```python
# Study each attribute, its associated values, and data type

for column in thyroid_data.columns:
    print("Attribute:", column)
    print("Pandas Data Type:", thyroid_data[column].dtype)
    print("Number of Unique Values:", thyroid_data[column].nunique())
    print("Associated Values:", thyroid_data[column].unique())
    print("-" * 50)
```

encoding scheme for the categorical attributes

```python
# Identify categorical attributes
categorical_columns = thyroid_data.select_dtypes(include=['object']).columns

print("Categorical Attributes:")
for column in categorical_columns:
    print(column, ":", thyroid_data[column].dropna().unique())
```

data range for numeric variables

```python
# Select numeric attributes
numeric_columns = thyroid_data.select_dtypes(include=[np.number]).columns

# Display minimum and maximum values for each numeric attribute
for column in numeric_columns:
    minimum_value = thyroid_data[column].min()
    maximum_value = thyroid_data[column].max()

    print("Attribute:", column)
    print("Minimum Value:", minimum_value)
    print("Maximum Value:", maximum_value)
    print("-" * 40)
```

presence of missing values

```python
# Calculate the number of missing values in each attribute
missing_values = thyroid_data.isnull().sum()

# Display missing values
print("Missing Values in Each Attribute:")
print(missing_values)
```

presence of outliers in data

```python
# Detect outliers in each numeric attribute using the IQR method

for column in numeric_columns:
    data = thyroid_data[column].dropna()

    # Calculate Q1 and Q3
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Calculate lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = data[(data < lower_bound) | (data > upper_bound)]

    print("Attribute:", column)
    print("Number of Outliers:", len(outliers))
    print("-" * 40)
```

mean var and std dev for numeric variables

```python
# Calculate mean, variance and standard deviation for numeric attributes

for column in numeric_columns:
    mean_value = thyroid_data[column].mean()
    variance_value = thyroid_data[column].var()
    std_value = thyroid_data[column].std()

    print("Attribute:", column)
    print("Mean:", mean_value)
    print("Variance:", variance_value)
    print("Standard Deviation:", std_value)
    print("-" * 40)
```
