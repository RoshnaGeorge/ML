A9: Data Normalization / Scaling

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load the thyroid dataset
thyroid_data = pd.read_excel(
    "Lab Session Data.xlsx",
    sheet_name="thyroid0387_UCI"
)

print("Dataset Shape:", thyroid_data.shape)
```

data prep

```python
# Replace '?' with NaN
thyroid_data = thyroid_data.mask(thyroid_data == '?', np.nan)

# Convert applicable columns to numeric
for column in thyroid_data.columns:
    converted = pd.to_numeric(thyroid_data[column], errors='coerce')

    if converted.notna().sum() == thyroid_data[column].notna().sum():
        thyroid_data[column] = converted

# Identify numeric attributes
numeric_columns = thyroid_data.select_dtypes(include=[np.number]).columns

print("Numeric Attributes:")
print(numeric_columns.tolist())
```

identify attributes that need normalization

```python
# Display the range of numeric attributes
for column in numeric_columns:
    print(
        column,
        "-> Min:", thyroid_data[column].min(),
        "Max:", thyroid_data[column].max()
    )
```

before normalization,handling missing values in numeric attributes.

```python
# Create a copy of the dataset
normalized_data = thyroid_data.copy()

# Fill missing numeric values using median
for column in numeric_columns:
    normalized_data[column] = normalized_data[column].fillna(
        normalized_data[column].median()
    )
```

min-max normalization

```python
# Create the Min-Max Scaler
scaler = MinMaxScaler()

# Apply normalization to numeric attributes
normalized_data[numeric_columns] = scaler.fit_transform(
    normalized_data[numeric_columns]
)

# Display normalized numeric data
print(normalized_data[numeric_columns].head())
```

verify normalization range

```python
# Check minimum and maximum values after normalization
for column in numeric_columns:
    print(
        column,
        "-> Min:", normalized_data[column].min(),
        "Max:", normalized_data[column].max()
    )
```
