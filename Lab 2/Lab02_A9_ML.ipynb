A9: Data Normalization / Scaling

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load ->thyroid dataset
thyroid_data = pd.read_excel(
    "Lab Session Data.xlsx",
    sheet_name="thyroid0387_UCI"
)
print("Dataset Shape:", thyroid_data.shape)
```

data prep

```python
thyroid_data = thyroid_data.mask(thyroid_data == '?', np.nan)
for col in thyroid_data.columns:
    conv = pd.to_numeric(thyroid_data[col], errors='coerce')

    if conv.notna().sum() == thyroid_data[col].notna().sum():
        thyroid_data[col] = conv
num_col = thyroid_data.select_dtypes(include=[np.number]).columns
print("Numeric Attributes:")
print(num_col.tolist())
```

identify attributes that need normalization

```python
for col in num_col:
    print(
        col,
        "-> Min:", thyroid_data[col].min(),
        "Max:", thyroid_data[col].max()
    )
```

before normalization,handling missing values in numeric attributes.

```python

normalized_data = thyroid_data.copy()
for col in num_col:
    normalized_data[col] = normalized_data[col].fillna(
        normalized_data[col].median()
    )
```

min-max normalization

```python

scaler = MinMaxScaler()

normalized_data[num_col] = scaler.fit_transform(
    normalized_data[num_col]
)

print(normalized_data[num_col].head())
```

verify normalization range

```python

for col in num_col:
    print(
        col,
        "-> Min:", normalized_data[col].min(),
        "Max:", normalized_data[col].max()
    )
```
