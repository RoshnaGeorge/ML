A8: Data Imputation


```python
import pandas as pd
import numpy as np

# Load ->thyroid dataset
thyroid_data = pd.read_excel(
    "Lab Session Data.xlsx",
    sheet_name="thyroid0387_UCI"
)

print("shape:", thyroid_data.shape)
```

```python

thyroid_data = thyroid_data.mask(thyroid_data == '?', np.nan)
print("missing val:")
print(thyroid_data.isnull().sum())
```

identify numeric and categorical attributes

```python

for col in thyroid_data.columns:
    converted = pd.to_numeric(thyroid_data[col], errors='coerce')

    if converted.notna().sum() == thyroid_data[col].notna().sum():
        thyroid_data[col] = converted

num_col = thyroid_data.select_dtypes(include=[np.number]).columns
cat_col = thyroid_data.select_dtypes(exclude=[np.number]).columns
print("numeric attributes:")
print(num_col.tolist())
print("\ncategorical attributes:")
print(cat_col.tolist())
```

numeric column for outliers using IQR

```python

for col in num_col:
    data = thyroid_data[col].dropna()

    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lb = Q1 - 1.5 * IQR
    ub = Q3 + 1.5 * IQR
    outliers = data[(data < lb) | (data > ub)]
    if len(outliers) > 0:
        fill_val = data.median()
        thyroid_data[col] = thyroid_data[col].fillna(fill_val)
        print(col, "-> Median Imputation")
    else:
        fill_val = data.mean()
        thyroid_data[col] = thyroid_data[col].fillna(fill_val)
        print(col, "-> Mean Imputation")
```

categorical

```python

for col in cat_col:
    if thyroid_data[col].isnull().any():
        mode_val = thyroid_data[col].mode()[0]
        thyroid_data[col] = thyroid_data[col].fillna(mode_val)
        print(col, "-> Mode Imputation")
```

verify missing values

```python

print("missing val:")
print(thyroid_data.isnull().sum())
print(
    "\ntotal missing values:",
    thyroid_data.isnull().sum().sum()
)
```
