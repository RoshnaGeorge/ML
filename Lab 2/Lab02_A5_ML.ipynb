A5: Similarity Measure Using Jaccard Coefficient and Simple Matching Coefficient

```python
import pandas as pd
import numpy as np

# Load -> thyroid dataset
thyroid_data = pd.read_excel(
    "Lab Session Data.xlsx",
    sheet_name="thyroid0387_UCI"
)

print(thyroid_data.head(2))
```

binary values

```python
# binary attributes
bin_col = []

for col in thyroid_data.columns:
    unique_val = set(thyroid_data[col].dropna().unique())

    if unique_val.issubset({'t', 'f'}) and len(unique_val) > 0:
        bin_col.append(col)

print("binary attributes:")
print(bin_col)

print("\nnumber of binary attributes:", len(bin_col))
```

first 2 binary observation vectors


```python
first_two = thyroid_data[bin_col].iloc[:2]
bin_data = first_two.replace({'t': 1, 'f': 0})

vector1 = bin_data.iloc[0].to_numpy()
vector2 = bin_data.iloc[1].to_numpy()

print("first binary ob vector:")
print(vector1)
print("\nsecond binary ob vector:")
print(vector2)
```

Calculate f11, f00, f01, and f10

```python
f11 = np.sum((vector1 == 1) & (vector2 == 1))
f00 = np.sum((vector1 == 0) & (vector2 == 0))
f01 = np.sum((vector1 == 0) & (vector2 == 1))
f10 = np.sum((vector1 == 1) & (vector2 == 0))
print("f11 =", f11)
print("f00 =", f00)
print("f01 =", f01)
print("f10 =", f10)
```

JC and SMC

```python

JC = f11 / (f01 + f10 + f11)
SMC = (f11 + f00) / (f00 + f01 + f10 + f11)

print("JC:", JC)
print("SMC:", SMC)
```
