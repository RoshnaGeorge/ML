A1: Purchase Data Analysis using Linear Algebra


```python
import pandas as pd
import numpy as np
```

```python
# Load the Purchase data worksheet
purchase_data = pd.read_excel("Lab Session Data.xlsx", sheet_name="Purchase data")

purchase_data.head()
```

```python
## Segregating the Data into Feature Matrix X and Output Vector y
X = purchase_data[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']].values

y = purchase_data['Payment (Rs)'].values

print("Feature Matrix X:")
print(X)

print("\nOutput Vector y:")
print(y)
```

1.What is the dimensionality of the vector space for this data?

The dimensionality of a vector space is the number of features in each observation. Here, each customer has 3 features: Candies (#), Mangoes (Kg), and Milk Packets (#). Therefore, the dimensionality of the vector space is 3.

```python
# dimensionality of the vector space
print("Shape of X:", X.shape)
print("Dimensionality of the vector space:", X.shape[1])
```

2.How many vectors exist in this vector space?

Each row in the feature matrix X represents one observation vector or one customer's purchase. Since the matrix has 10 rows, there are 10 observation vectors, and each vector has 3 dimensions.

```python
# Find the number of observation vectors
number_of_vectors = X.shape[0]

print("Number of observation vectors:", number_of_vectors)
```

3.What is the rank of Matrix A? How to calculate rank of a matrix? Calculate the
rank of the feature matrix.



*   The rank of a matrix is the maximum number of linearly independent rows or columns in the matrix
*   The rank can be calculated manually by converting the matrix into Row Echelon Form and counting the number of non-zero rows.






4.Calculate the rank of the feature matrix with numpy package

```python
# Calculate the rank of the feature matrix X
rank_X = np.linalg.matrix_rank(X)

print("Rank of feature matrix X:", rank_X)
```

The calculated rank is 3. This means the three feature columns are linearly independent.

5.Using Pseudo-Inverse find the cost of each product available for sale.

```python
# Cost of Each Product Using Pseudo-Inverse
X_pseudo_inverse = np.linalg.pinv(X)

product_costs = X_pseudo_inverse @ y

print("Cost of one Candy: Rs.", product_costs[0])
print("Cost of 1 kg Mango: Rs.", product_costs[1])
print("Cost of one Milk Packet: Rs.", product_costs[2])
```

### Conclusion

* The Purchase Data contains 10 observation vectors with 3 features.
* The dimensionality of the vector space is 3.
* The rank of the feature matrix X is 3, indicating that the three feature columns are linearly independent.
* Using the pseudo-inverse method, the individual product costs were calculated.
   * The cost of one Candy: ~ Rs. 1.
   * The cost of 1 kg of Mango: ~ Rs. 55.
   * The cost of one Milk Packet: Rs. 18.
