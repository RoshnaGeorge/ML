A5: Similarity Measure Using Jaccard Coefficient and Simple Matching Coefficient

```python
import pandas as pd
import numpy as np

# Load the thyroid dataset
thyroid_data = pd.read_excel(
    "Lab Session Data.xlsx",
    sheet_name="thyroid0387_UCI"
)

# Display the first two observations
print(thyroid_data.head(2))
```

binary values

```python
# Identify binary attributes containing only 't' and 'f'
binary_columns = []

for column in thyroid_data.columns:
    unique_values = set(thyroid_data[column].dropna().unique())

    if unique_values.issubset({'t', 'f'}) and len(unique_values) > 0:
        binary_columns.append(column)

print("Binary Attributes:")
print(binary_columns)

print("\nNumber of Binary Attributes:", len(binary_columns))
```

first 2 binary observation vectors


```python
# Select the first two observations using only binary attributes
first_two = thyroid_data[binary_columns].iloc[:2]

# Convert 't' and 'f' into 1 and 0
binary_data = first_two.replace({'t': 1, 'f': 0})

# Extract the two observation vectors
vector1 = binary_data.iloc[0].to_numpy()
vector2 = binary_data.iloc[1].to_numpy()

# Display the vectors
print("First Binary Observation Vector:")
print(vector1)

print("\nSecond Binary Observation Vector:")
print(vector2)
```

Calculate f11, f00, f01, and f10

```python
# Calculate f11, f00, f01 and f10

f11 = np.sum((vector1 == 1) & (vector2 == 1))
f00 = np.sum((vector1 == 0) & (vector2 == 0))
f01 = np.sum((vector1 == 0) & (vector2 == 1))
f10 = np.sum((vector1 == 1) & (vector2 == 0))

# Display the results
print("f11 =", f11)
print("f00 =", f00)
print("f01 =", f01)
print("f10 =", f10)
```

JC and SMC

```python
# Calculate Jaccard Coefficient (JC)
JC = f11 / (f01 + f10 + f11)

# Calculate Simple Matching Coefficient (SMC)
SMC = (f11 + f00) / (f00 + f01 + f10 + f11)

# Display the results
print("Jaccard Coefficient (JC):", JC)
print("Simple Matching Coefficient (SMC):", SMC)
```
